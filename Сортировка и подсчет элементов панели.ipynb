```python
import pandas as pd
import os

# Установим рабочую директорию
os.chdir(r'C:\Users\user\Desktop\template_programm_divided')

# Путь к исходному файлу
input_file = 'armatura_output_10CWA62GA201.xlsx'

# Загрузка данных из Excel
df = pd.read_excel(input_file)

# Функция для извлечения букв и чисел из первого столбца
def extract_sort_key(coordinate):
    # Разделяем строку на буквы и числа
    parts = coordinate.split('-')
    if len(parts) == 2:
        return parts[0], int(parts[1])  # Возвращаем буквы и числа как кортеж
    return coordinate, 0  # Если формат неверный, возвращаем исходное значение

# Добавляем временный столбец для сортировки
df['SortKey'] = df['Coordinate'].apply(extract_sort_key)

# Сортируем DataFrame по временному столбцу
df = df.sort_values(by='SortKey')

# Удаляем временный столбец
df = df.drop(columns=['SortKey'])

# Подсчитываем количество уникальных элементов в третьем столбце
third_column_name = df.columns[2]  # Предположим, что третий столбец имеет индекс 2
unique_counts = df[third_column_name].value_counts().reset_index()
unique_counts.columns = ['UniqueValue', 'Count']

# Создаем пустую строку для разделения основных данных и подсчета уникальных элементов
separator = pd.DataFrame([[''] * len(df.columns)], columns=df.columns)

# Объединяем оригинальный DataFrame, разделитель и DataFrame с уникальными значениями
output_df = pd.concat([df, separator, unique_counts], ignore_index=True)

# Формируем путь для сохранения файла в той же папке
output_file = os.path.join(os.path.dirname(input_file), 'sorted_' + os.path.basename(input_file))

# Сохраняем отсортированный DataFrame в новый Excel-файл
output_df.to_excel(output_file, index=False)
print(f"Отсортированный файл с подсчетом уникальных элементов сохранен как: {output_file}")

# Выводим результаты подсчета уникальных элементов в консоль (опционально)
print(f"Подсчет уникальных элементов в столбце '{third_column_name}':")
print(unique_counts)
```
